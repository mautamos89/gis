```python
print(u'título: Geocodificación completa de lista de direcciones con coordenadas en grados decimales (WGS84)'.upper())
print('autor:  mauricio tabares mosquera'.upper())
print('fecha:  2023-02-18'.upper())
print(u'instalando librerías y módulos'.upper())
#!pip install geocoder
import geocoder as gc, pandas as pd, matplotlib.pyplot as plt, time
print(u'geocodificación iniciada'.upper())
#time.clock = time.time
PStart = time.time() # Start clock
Data = '/content/drive/MyDrive/Procesamiento/Gecode_SUI_20230218.xlsx' # Data variable
Sheet = sheet_name='G2'
DatTxt = '/content/G2.txt'
TabExc = pd.read_excel(Data,Sheet) # Read excel file with pandas library
f=open(DatTxt,'w',encoding='utf-8') #Text file with shortened url per line
i = 0
while i < len(TabExc): # Iterate and concatenate address with latitude and longitude
  Id = TabExc.loc[i][0]
  Dir = TabExc.loc[i][1]
  Ciu = TabExc.loc[i][2]
  #Dto = TabExc.loc[i][3]
  Pai = TabExc.loc[i][4]
  FullDir = Dir + ', ' + Ciu + ', ' + Pai
  LocArc = gc.arcgis(FullDir) # Servicio de geocidificación de Esri
  LatLon = LocArc.latlng
  i +=1
  print(i, ';' ,str(int(Id)), ';' ,FullDir, ';', LocArc, ';', 'Latitud:', LatLon[0],'; Longitud:', LatLon[1])
  f.write(str(int(Id)) + ';' +str(FullDir)+ ';'+ str(LocArc)+ ';'+ 'Latitud:'+ str(LatLon[0])+'; Longitud:'+ str(LatLon[1])+'\n')
  time.sleep(.1)
f.close()
PEnd = int((time.time()-PStart)/60) # Execution time
print(u'geocodificación terminada. '.upper() + u'Tiempo de ejecución: {} minutos'.format(PEnd).upper())
```
