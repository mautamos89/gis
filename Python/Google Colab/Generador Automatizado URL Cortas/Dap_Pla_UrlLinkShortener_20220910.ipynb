```python
# TRANSFORMS A SET OF URLS TO TINY-URLS (SHORTENED URLS)
# Input: UrlFile.txt with one URL per line
# Output: TinyUrlFile.txt, with the original URL and the shortened one

print('SCRIPT STARTED')

from __future__ import with_statement                                                            
import contextlib 
try: 
    from urllib.parse import urlencode           
except ImportError: 
    from urllib import urlencode 
try: 
    from urllib.request import urlopen 
except ImportError: 
    from urllib2 import urlopen 
  
#import sys 
import time
  
def make_tiny(url): 
    request_url = ('http://tinyurl.com/api-create.php?' + urlencode({'url':url}))     
    with contextlib.closing(urlopen(request_url)) as response:                       
        return response.read().decode('utf-8 ')                                       

con = open('PlaUrl_Long.txt', 'r',encoding='latin1').readlines() #Text file with original url per line

urls = []
for i in con: urls.append(i.replace('\n', '').replace('\t',''))

f=open('PlaUrl_Short.txt','w',encoding='utf-8') #Text file with shortened url per line

Count = 0
for i in urls:
    tiny='NA'
    #print(i)
    if (len(i)>3): tiny=make_tiny(i)   
    f.write(i+'\t'+tiny+'\n')
    time.sleep(.3)
    Count +=1
    print(str(Count) + ';' +  str(tiny))
    
f.close()

print('SCRIPT FINISHED')
```
